## HYPOTHESIS TESTING

```python
### Cost Model 
### W= 1000 + 5X
### w= weekely  operating cost, X=units produced 
### sample mean cost = 3050
### sample size(n)=25
### mean unit produced (μ)=600
### Standard deviation (σ) =25 
```

### Step 1: State Hypotheses
### H0=claim is false 
### H1=Claim is true 
### H0:μ=1000+5X  
### H1:μ>1000+5X  


### Step 2: Calculate the Test Statistic
### μ=1000+3000=4000
### σ=5×25=125
### Z=−38
### At 5% significance level (α = 0.05):
### Right-tailed test → Critical Z = +1.645

### Step 3: Conclusion (VERY IMPORTANT)
### There is NO statistical evidence to support the claim that:
### Weekly operating costs are higher than expected.
### Observed cost (₹3050) is much lower than expected (₹4000)

```python
# Python Code
import math
x_bar=3050
mu=4000
sigma=125
n=25
z= (x_bar - mu)/(sigma/math.sqrt(n))
print("Z value: ", z)
critical_z = 1.645
if z > critical_z:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```
