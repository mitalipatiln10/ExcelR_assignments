# Exercise 1: Prime Numbers

```python
import math 
num = int(input("Enter Prime Number: ")) 
if num > 1 :
        for i in range(2, int(math.sqrt(num)) + 1):
             if num % i == 0:
                 print(num,"is not a Prime Number")
                 break

        else:
            print(num," is prime number")

else:
    print(num,"is not a prime number")
    
    
    
    


```

# Exercise 2: Product of Random Numbers

```python
import random
a = random.randint(1,10)
b = random.randint(1,10)

ans = int(input(f"What is {a} x {b}"))

if ans == a*b:
    print("Correct!")
else:
    print("Wrong! Correct ans is" , a*b)
    
          
```

# Exercise 3: Squares of Even/Odd Numbers

```python
# even number
for num in range (100,201):
    if num % 2 == 0 :
        print(num,"square is: ",num**2)
```

```python
# odd number
for num in range (100,201):
    if num % 2 != 0 :
        print(num,"square is: ",num**2)
```

# Exercise 4:

```python
room_charge_per_day = 2000
days = 4 
doctor_fees = 1500
lab_charges_fees = 300
test = 3
medicines_charges = 2400
# calculation
room_total = room_charge_per_day * days
lab_charges = lab_charges_fees * test
total_bill = room_total + doctor_fees + lab_charges + medicines_charges
# Apply discount if bill exceeds 10000
if total_bill > 10000:
    discount = total_bill *0.10
else:
    discount = 0 
final_bill = total_bill - discount
#Output
print("Room Charges:", room_total)
print("Doctor Fee:", doctor_fees)
print("Lab Test Charges:", lab_charges)
print("Medicine Charges:", medicines_charges)
print("Total Bill Before Discount:", total_bill)
print("Discount:", discount)
print("Final Bill:", final_bill)
```
